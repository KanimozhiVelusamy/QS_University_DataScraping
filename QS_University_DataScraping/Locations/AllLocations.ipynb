```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Load university URLs from the CSV file
df = pd.read_csv(r'C:\Users\kanim\Downloads\universities_with_no_650.csv')

def extract_locations(url):
    page = requests.get(url).text
    soup = BeautifulSoup(page, 'html.parser')
    
    # Extract primary location
    primary_location_tag = soup.find('li', {'class': 'campus-locations-links datapoints_js active'})
    primary_location = ""
    if primary_location_tag is not None:
        span_tag = primary_location_tag.find('span')
        if span_tag is not None:
            primary_location = span_tag.text.strip()
    
    # Extract other campus locations
    other_locations_tags = soup.find_all('li', {'class': 'campus-locations-links datapoints_js'})
    other_locations = [location.text.strip() for location in other_locations_tags]
    
    return primary_location, other_locations

# Lists to store data
universities = []
primary_locations = []
other_locations = []

# Iterate over each university URL and extract locations
for index, row in df.iterrows():
    university_name = row['University']
    university_url = row['URL']
    
    print(f"University: {university_name}")
    print(f"URL: {university_url}")
    
    primary_location, other_locations_list = extract_locations(university_url)
    print(f"Primary Location: {primary_location}")
    print("Other Locations:")
    for loc in other_locations_list:
        print(loc)
    print()
    
    universities.append(university_name)
    primary_locations.append(primary_location)
    other_locations.append(other_locations_list)

# Create DataFrame to store the data
data = pd.DataFrame({
    'University': universities,
    'Primary Location': primary_locations,
    'Other Locations': other_locations
})

# Save data to CSV
data.to_csv('university_locations_7.csv', index=False)
print("Location data saved to 'university_locations_7.csv'")
```
